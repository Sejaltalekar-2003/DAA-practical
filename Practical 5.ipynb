```python
def is_safeplace(board,row,col,n):
    for i in range(row):
        if board[i][col]==1:
            return  False
        
    for i ,j in zip ( range(row,-1,-1),range(col,-1,-1) ):
        if board[i][j]==1:
            return False
        
    for i ,j in zip ( range(row,-1,-1),range(col,n)):
        if board[i][j]==1:
            return False
        
    return True


def solve_queen_util(board,row,n,solutions):
    if row==n:
        solution=[]
        for i in range (n):
            solution.append(board[i][:])
        solutions.append(solution)
        return 
    
    for col in range(n):
        if is_safeplace(board,row,col,n):
            board[row][col] = 1
            solve_queen_util(board,row+1,n,solutions)
            board[row][col] = 0
            
def solve_queens(n):
    board=[[0 for _ in range(n)]for _ in range(n)]
    solutions=[]
    solve_queen_util(board,0,n,solutions)
    return solutions

def print_solution(solutions):
    count = 1
    for solution in solutions:
        print(f"Solution {count}:")
        for row in solution:
            print(row)
        print()
        count += 1
        
        
n = int(input("Enter the n value for n quuen:"))
if n < 1:
    print("Invalid input")
else:
    solutions = solve_queens(n)
    if solutions:
        print_solution(solutions)
    else:
        print(f"There is no solution for{n}queen:")
```
