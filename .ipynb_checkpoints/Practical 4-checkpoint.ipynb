```python
def knapsack(max_capacity, weights, profits, num_item):
    # Initialize the DP matrix with zero values
    K = [[0] * (max_capacity + 1) for i in range(num_item + 1)]
    
    # Fill the DP matrix
    for i in range(num_item + 1):
        for w in range(max_capacity + 1):
            if i == 0 or w == 0:
                K[i][w] = 0
            elif weights[i - 1] <= w:
                K[i][w] = max(profits[i - 1] + K[i - 1][w - weights[i - 1]], K[i - 1][w])
            else:
                K[i][w] = K[i - 1][w]
    
    # Display the DP matrix
    print("DP Matrix:")
    for row in K:
        print(row)
    
    # Determine the maximum profit
    max_profit = K[num_item][max_capacity]
    selected_items = []
    remaining_capacity = max_capacity
    
    # Trace back to find selected items
    for i in range(num_item, 0, -1):
        if K[i][remaining_capacity] != K[i - 1][remaining_capacity]:
            selected_items.append(i)
            remaining_capacity -= weights[i - 1]
    
    return max_profit, selected_items

# User input
num_item = int(input("Enter the number of items: "))
profits = []
weights = []

# Collect weights and profits for each item
for i in range(num_item):
    weight = int(input(f"Enter the weight of item {i + 1}: "))
    profit = int(input(f"Enter the profit of item {i + 1}: "))
    weights.append(weight)
    profits.append(profit)

max_capacity = int(input("Enter Maximum Capacity: "))

# Call the knapsack function
total_profit, selected_items = knapsack(max_capacity, weights, profits, num_item)

# Display the results
print("\nAll items with their weights and profits:")
print("Items (Weight, Profit)")
for i in range(num_item):
    print(f"Item {i + 1}: Weight = {weights[i]}, Profit = {profits[i]}")

print("\nItems included to maximize profit:")
print("Item | Weight | Profit")
for item in selected_items[::-1]:  # Reverse to show in original order
    print(f"{item}     | {weights[item - 1]}     | {profits[item - 1]}")
```
